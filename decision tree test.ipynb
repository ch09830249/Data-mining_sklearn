```python
#我們可以用export_graphviz將樹導出為Graphviz格式，在命令行輸入dot -Tpdf ID3.dot -o ID3.pdf生成決策樹的PDF可視化文件
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
#用此模組
import graphviz

data =  pd.read_csv("train_and_test2.csv")
record=[]
record_list=[]
for i in range(0,data.shape[0]):
    for j in range(0,data.shape[1]):
        if j==0:
            if data.values[i,j] <= 35:
                record_list.append(0)    # young
            elif data.values[i,j] <= 65:
                record_list.append(1)    # middle
            else:
                record_list.append(2)    # old
        elif j==2:
            if data.values[i,j] == 0:
                record_list.append(0)    # no
            else:
                record_list.append(1)    # yes
        else:
            record_list.append(data.values[i,j])
    record.append([(record_list[k]) for k in range(0,len(record_list))])
    record_list=[]

X = np.array(record)[:, :3]
Y = np.array(record)[:, 3]
print(X)
print(Y)

# Using decision tree for classification
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X, Y)
print("clf: " + str(clf)) 

# Visualize model
with open("ID3.dot", 'w') as f:
    f = tree.export_graphviz(clf, feature_names=['age','sex','sibsp'], out_file=f)
    
f1 = tree.export_graphviz(clf, feature_names=['age','sex','sibsp'], out_file=f)
graph = graphviz.Source(f1)
graph.render(r'D:\ID3')
```
